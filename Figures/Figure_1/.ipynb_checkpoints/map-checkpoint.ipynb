```python
import geopandas as gpd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.lines as mlines
from matplotlib.colors import ListedColormap
import re, os, glob
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt


mpl.rcParams['font.family'] = 'sans-serif'

df=pd.read_csv("epidata.csv")
df=pd.DataFrame(df["Total Case Count"].groupby(df["State"]).sum()).reset_index()

rn={'Bauchi ': 'Bauchi',
    'Cross Rivers': 'Cross River',
    'FCT':'Federal Capital Territory',
    "Fct":"Federal Capital Territory" }

df["State"]=[rn[x] if x in rn.keys() else x for x in df["State"].to_list() ]
ss=dict(zip(df["State"], df["Total Case Count"]))

nigeria_geopolitical_zones = {
    'Northwest': ['Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Sokoto', 'Zamfara'],
    'Northeast': ['Adamawa', 'Bauchi', 'Borno', 'Gombe', 'Taraba', 'Yobe'],
    'North Central': ['Benue', 'Kogi', 'Kwara', 'Nasarawa', 'Niger', 'Plateau', 'FCT (Abuja)'],
    'Southwest': ['Ekiti', 'Lagos', 'Ogun', 'Ondo', 'Osun', 'Oyo'],
    'Southeast': ['Abia', 'Anambra', 'Ebonyi', 'Enugu', 'Imo'],
    'South-South': ['Akwa Ibom', 'Bayelsa', 'Cross River', 'Delta', 'Edo', 'Rivers']}

reversed_nigeria_zones = {state: zone for zone, states in nigeria_geopolitical_zones.items() for state in states}

dd={'North Central': "#960018",
  'Northeast': "#FE5A1D",
  'Northwest': "#FFD700",
  'South-South': "#16166B",
  'Southeast': "#0070BB",
  'Southwest': "#138808",
  'Not_specified': "#C0C0C0"}


RN=dict(zip(['North Central', 'Northeast', 'Northwest', 'South-South', 'Southeast', 'Southwest', 'Not_specified'],
           ['North Central', 'North East', 'North West', 'South South', 'South East', 'South West', 'Not_specified']))

geojson_path = "Nigeria.geojson"
nigeria_states = gpd.read_file(geojson_path)
nigeria_states["Count"]=[ss[x] if x in ss.keys() else 0 for x in  nigeria_states["admin1Name"].to_list()]
nigeria_states["Region"]=nigeria_states["admin1Name"].map(reversed_nigeria_zones)
nigeria_states["CC"]=nigeria_states["Region"].map(dd)
nigeria_states["Region"].loc[nigeria_states["CC"].isna()]="North Central"
nigeria_states["CC"].loc[nigeria_states["CC"].isna()]="#960018"
```

```python

fig, ax = plt.subplots(1, 1, figsize=(12, 8))

custom_colors = ['#960018', '#FE5A1D', '#FFD700', '#16166B', '#0070BB', '#138808']
custom_cmap = ListedColormap(custom_colors)

nigeria_states.plot(column='Region', cmap=custom_cmap, linewidth=0.75, ax=ax, 
                    edgecolor="black", alpha=0.75)


ax.set_yticklabels([])
ax.set_xticklabels([])

ax.tick_params(axis='y',size=0)
ax.tick_params(axis='x',size=0)
[ax.spines[loc].set_visible(False) for loc in ['top','right','left','bottom']]

target_row = nigeria_states[nigeria_states["admin1Name"]=="Abia"]
target_row.plot(ax=ax, linewidth=2, edgecolor='#FF0800')
target_row = nigeria_states[nigeria_states["admin1Name"]=="Rivers"]
target_row.plot(ax=ax, linewidth=2, edgecolor='#7C0902')

uu=13.5
yy=10
for k,v in dd.items():
    if k != "Not_specified":
        ax.scatter(uu,yy,s=400,color=v,zorder=400, edgecolor='black')
        ax.text(uu+0.3, yy-0.175, RN[k], size=30)
        # uu=uu+0.005
        yy=yy-0.75    

ax.scatter(uu,yy,s=400,color="#FF0800",zorder=400, edgecolor='black', label='Line', marker='+')
ax.text(uu+0.3, yy-0.175, "Abia State", size=30)    

ax.scatter(uu,yy-0.5,s=400,color="#7C0902",zorder=400, edgecolor='black', label='Line', marker='+')
ax.text(uu+0.3, yy-0.175-0.75, "Rivers State", size=30)    

#ax.set_title('B', fontsize=25, loc='left')


#plt.show()

plt.savefig("Map.png",bbox_inches='tight', dpi=1000)   
```
